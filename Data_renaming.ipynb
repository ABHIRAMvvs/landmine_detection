```python
import os
import pandas as pd

# Function to rename images and save their previous names in an Excel file
def rename_images_and_save_excel(root_dir):
    # Initialize list to store data for DataFrame
    data = []

    # Initialize counter for new image names
    count = 1

    # Traverse through the directories
    for date_dir in os.listdir(root_dir):
        date_dir_path = os.path.join(root_dir, date_dir)
        for subdir in os.listdir(date_dir_path):
            subdir_path = os.path.join(date_dir_path, subdir)
            if os.path.isdir(subdir_path):  # Check if it's a directory
                for zone_dir in os.listdir(subdir_path):
                    zone_dir_path = os.path.join(subdir_path, zone_dir)
                    for image_file in os.listdir(zone_dir_path):
                        if image_file.endswith('.jpg') or image_file.endswith('.jpeg') or image_file.endswith('.png'):
                            # Construct new image name
                            new_name = f"{count}.jpg"
                            count += 1

                            # Rename image file
                            old_path = os.path.join(zone_dir_path, image_file)
                            new_path = os.path.join(zone_dir_path, new_name)
                            os.rename(old_path, new_path)

                            # Add data for DataFrame
                            data.append({'Previous Name': image_file, 'New Name': new_name})

    # Create DataFrame from the collected data
    df = pd.DataFrame(data)

    # Save DataFrame to Excel
    excel_file = os.path.join(root_dir, 'image_names.xlsx')
    df.to_excel(excel_file, index=False)

# Define the root directory of the dataset
root_directory = r"C:\Users\Saiab\Desktop\Land mines project\temp_dataset_1"

# Call the function to rename images and save their previous names in an Excel file
rename_images_and_save_excel(root_directory)
```

```python
import os

def find_images_below_altitude(root_dir, max_altitude=2):
    # Initialize a list to store paths of images below the specified altitude
    images_below_altitude = []

    # Check if the root directory exists
    if not os.path.isdir(root_dir):
        print(f"Error: '{root_dir}' is not a valid directory.")
        return images_below_altitude

    # Traverse through the directories
    for date_dir in os.listdir(root_dir):
        date_dir_path = os.path.join(root_dir, date_dir)
        
        # Check if it's a directory
        if not os.path.isdir(date_dir_path):
            continue
        
        for subdir in os.listdir(date_dir_path):
            subdir_path = os.path.join(date_dir_path, subdir)
            
            # Check if it's a directory
            if not os.path.isdir(subdir_path):
                continue
            
            for zone_dir in os.listdir(subdir_path):
                zone_dir_path = os.path.join(subdir_path, zone_dir)
                
                # Check if it's a directory
                if not os.path.isdir(zone_dir_path):
                    continue
                
                for image_file in os.listdir(zone_dir_path):
                    if image_file.endswith('.jpg') or image_file.endswith('.jpeg') or image_file.endswith('.png'):
                        # Extract altitude information from the filename
                        parts = image_file.split('__')
                        if len(parts) >= 2:
                            altitude_str = parts[1].split('m')[0]
                            try:
                                altitude = float(altitude_str)
                            except ValueError:
                                altitude = None
                            if altitude is not None and altitude < max_altitude:
                                images_below_altitude.append(os.path.join(zone_dir_path, image_file))

    return images_below_altitude

# Define the root directory of the dataset
root_directory = r"C:\Users\Saiab\Desktop\Land mines project\Dataset"

# Find images below 0.2 meters altitude
images_below_0_2m = find_images_below_altitude(root_directory, max_altitude=1)

# Print the paths of images below 0.2 meters altitude
count = 0
for image_path in images_below_0_2m:
    count+=1
    print(image_path)
print(count)
```

```python
import os

def find_images_below_altitude(root_dir, max_altitude=2):
    # Initialize a list to store paths of images below the specified altitude
    images_below_altitude = []

    # Define the subfolders to search within
    subfolders = ['jpg', 'r_jpg']

    # Traverse through the subfolders
    for subfolder in subfolders:
        subfolder_path = os.path.join(root_dir, subfolder)
        if not os.path.isdir(subfolder_path):
            print(f"Error: '{subfolder}' folder does not exist.")
            continue

        # Traverse through the zone folders
        for zone_folder in os.listdir(subfolder_path):
            zone_folder_path = os.path.join(subfolder_path, zone_folder)
            if not os.path.isdir(zone_folder_path):
                continue

            # Iterate over the image files in the zone folder
            for image_file in os.listdir(zone_folder_path):
                if image_file.endswith('.jpg') or image_file.endswith('.jpeg') or image_file.endswith('.png'):
                    # Extract altitude information from the filename
                    parts = image_file.split('__')
                    if len(parts) >= 2:
                        altitude_str = parts[0].split('_')[-1].replace('m', '')
                        try:
                            altitude = float(altitude_str)
                        except ValueError:
                            altitude = None
                        if altitude is not None and altitude < max_altitude:
                            images_below_altitude.append(os.path.join(zone_folder_path, image_file))

    return images_below_altitude

# Define the root directory of the dataset
root_directory = r"C:\Users\Saiab\Desktop\Land mines project\Dataset\03_03_2020"

# Find images below 0.5 meters altitude
images_below_0_5m = find_images_below_altitude(root_directory, max_altitude=0.5)

# Print the paths of images below 0.5 meters altitude
for image_path in images_below_0_5m:
    print(image_path)
```
